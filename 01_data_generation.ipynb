# 01 - Data Generation

This notebook generates synthetic employee data for the Hamburg metropolitan area. We'll create realistic employee profiles with German names and addresses distributed across Hamburg and surrounding cities like Norderstedt, Harburg, and Wedel.

```python
# Import the libraries we need for generating synthetic data
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
# Set up Faker with German locale to get realistic German names and addresses
fake = Faker('de_DE')
Faker.seed(42)  # Fixed seed so we get the same results every time
```

```python
# Define Hamburg districts and surrounding cities with their postal codes
# This gives us realistic geographic distribution for our synthetic employees
hamburg_areas = {
    'Hamburg': ['20095', '20097', '20099', '20144', '20146', '20148', '20149', 
                '20251', '20253', '20255', '20259', '20354', '20355', '20357',
                '20535', '20537', '20539', '21029', '21031', '21033', '21035',
                '21037', '21039', '22041', '22043', '22045', '22047', '22049',
                '22081', '22083', '22085', '22087', '22089', '22113', '22115',
                '22117', '22119', '22143', '22145', '22147', '22149', '22297',
                '22299', '22301', '22303', '22305', '22307', '22309', '22391',
                '22393', '22395', '22397', '22415', '22417', '22419', '22453',
                '22455', '22457', '22523', '22525', '22527', '22529', '22547',
                '22549', '22559', '22587', '22589'],
    'Norderstedt': ['22844', '22846', '22848', '22850', '22851', '22852', '22854', '22855', '22856'],
    'Harburg': ['21073', '21075', '21077', '21079', '21107', '21109'],
    'Ahrensburg': ['22626', '22627', '22629'],
    'Wedel': ['22880'],
    'Buxtehude': ['21614', '21682'],
}
```

```python
# Function to generate synthetic employee data
def generate_synthetic_employees_data(num_employees=200):
    """Generate synthetic employee data for Hamburg area"""
    
    employees = []
    
    for i in range(num_employees):
        # Pick a random city/area from our defined list
        city = random.choice(list(hamburg_areas.keys()))
        postal_code = random.choice(hamburg_areas[city])
        
        # Generate realistic address components using Faker
        street_name = fake.street_name()
        street_number = fake.building_number()
        full_address = f"{street_name} {street_number}, {postal_code} {city}"
        
        # Create employee record
        employee = {
            'employee_id': f'EMP{i+1:03d}',
            'name': fake.name(),
            'address': full_address,
            'city': city,
            'postal_code': postal_code,
        }
        
        employees.append(employee)
    
    return pd.DataFrame(employees)

# Generate 200 synthetic employees
employee_df = generate_synthetic_employees_data(200)
employee_df.head()
```

```python
# Check how our synthetic employees are distributed across cities
print(employee_df['city'].value_counts())

# Look at postal code distribution too
print(employee_df['postal_code'].value_counts().head(10))

# Show a sample of our generated data
print(employee_df.head(10))
```

```python
# Save our synthetic employee data to CSV for use in other notebooks
employee_df.to_csv('synthetic_employees.csv', index=False)
print(f"Generated {len(employee_df)} synthetic employees and saved to synthetic_employees.csv")
```
